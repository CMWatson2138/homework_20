Import modules and text

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import gutenberg
```

```python
paradise_lost = nltk.corpus.gutenberg.words('milton-paradise.txt')
```

```python
paradise_lost
```

Tokenized

```python
text_sentences = nltk.corpus.gutenberg.sents('milton-paradise.txt')
print(text_sentences[:5])
```

Changing all to lowercase

```python
lower_text_sentences=[]
for sentence in text_sentences:
    for w in sentence:
        lower_text_sentences.append(w.lower())
lower_text_sentences
```

Removing stop words

```python
from nltk.corpus import stopwords

sw=stopwords.words('english')
```

```python
clean_tokens=lower_text_sentences[:]
for token in lower_text_sentences:
    if token in sw:
        clean_tokens.remove(token)
```

Remove nonalphas

```python
import string
string.punctuation
```

```python
cleaner_tokens=clean_tokens[:]
for token in clean_tokens:
    if token in string.punctuation:
        cleaner_tokens.remove(token)  
```

Remove nonsense

```python
nonsense_words=[ 'u', 'st']
cleanest_tokens=[]
for token in cleaner_tokens:
    if token not in nonsense_words:
        cleanest_tokens.append(token)
        
print(cleanest_tokens)    
```

Counting words

```python
wordfreq = []
for w in cleanest_tokens:
    wordfreq.append(cleanest_tokens.count(w))

pairs= str(list(zip(cleanest_tokens, wordfreq)))
```

```python
print(pairs)
```

```python
from nltk.stem import WordNetLemmatizer

wordnet_lemmatizer = WordNetLemmatizer()

lemmatized = [wordnet_lemmatizer.lemmatize(w) for w in cleanest_tokens]

bow = Counter(lemmatized)

bow1=bow.most_common(20)
print(bow1)
```
